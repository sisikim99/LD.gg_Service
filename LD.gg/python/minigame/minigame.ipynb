```python
# df 만들기
def mini_df(mini_data) :
    lst_mini = []
    for i in range(len(mini_data)) :
        for j in range(len(mini_data[i]['info']['participants'])) :
            lst_participant = []
            lst_participant.append(mini_data[i]['info']['participants'][j]['win'])
            lst_participant.append(str("KR_") + str(mini_data[i]['info']['gameId']))
            lst_participant.append(mini_data[i]['info']['participants'][j]['participantId'])
            lst_participant.append(mini_data[i]['info']['gameDuration'])
            lst_participant.append(mini_data[i]['info']['participants'][j]['teamPosition'])
            lst_participant.append(mini_data[i]['info']['participants'][j]['teamId'])
            lst_participant.append(mini_data[i]['info']['participants'][j]['championId'])
            perkString = ""
            for k in range(len(mini_data[i]['info']['participants'][j]['perks']['styles'])) :
                for l in range(len(mini_data[i]['info']['participants'][j]['perks']['styles'][k]['selections'])) :
                    perkString += str(mini_data[i]['info']['participants'][j]['perks']['styles'][k]['selections'][l]['perk'])
                    if(k == len(mini_data[i]['info']['participants'][j]['perks']['styles'])-1) :
                        if(k != l) :
                            perkString += '|'
                    else :
                        perkString += '|'
            lst_participant.append(perkString);
            spellString = ""
            spellString = str(mini_data[i]['info']['participants'][j]['summoner1Id']) + '|' + str(mini_data[i]['info']['participants'][j]['summoner2Id'])
            lst_participant.append(spellString)
            lst_mini.append(lst_participant)
    col_lst = ['minigame_result', 'match_id', 'participantId', 'mini_duration', 'mini_lane', 'mini_team', 'champion_id', 'mini_rune', 'mini_spell']
    df = pd.DataFrame(lst_mini, columns=col_lst)
    return df
```

```python
def time_line_data(time__):
    time_line = {}
    for i, frame in enumerate(time__['info']['frames']):
        result = {'events': [], 'participantFrames': []}
        events = []
        for event in frame['events']:
            if event['type'] == 'SKILL_LEVEL_UP' or event['type'] == 'ITEM_PURCHASED' or event[
                'type'] == 'BUILDING_KILL':
                events.append(event)
        result['events'].extend(events)
        for participant in range(len(frame['participantFrames'])):
            participant_dict = {
                'participantId': frame['participantFrames'][f'{participant + 1}']['participantId'],
                'level': frame['participantFrames'][f'{participant + 1}']['level'],
                'totalGold': frame['participantFrames'][f'{participant + 1}']['totalGold']
            }
            result['participantFrames'].append(participant_dict)
        time_line[i] = result
    return time_line
```

```python
def time_line_data(parti, time_d, mini_d) :
    lst_parti = []
    kill = 0
    assist = 0
    death = 0
    ward_placed = 0
    tower_kill = 0
    skill_tree = ''
    item_id = ''
    match_id = ''


    for i in range(len(time_d['info']['frames'])) :
        lst_min_parti = []

        # mini_t_time,int -- 제공 시간대
        mini_t_time = i;
        # mini_t_level,int -- 레벨
        mini_t_level = time_d['info']['frames'][i]['participantFrames']['1']['level']
        # time_line 추출.
        # KDA
        for j in range(len(time_d['info']['frames'][i]['events'])) :
            # KILL, ASSIST, DEATH
            if 'CHAMPION_KILL' in time_d['info']['frames'][i]['events'][j]['type'] :
                if 'assistingParticipantIds' in time_d['info']['frames'][i]['events'][j].keys() :
                    if parti in time_d['info']['frames'][i]['events'][j]['assistingParticipantIds'] :
                        assist += 1
                if parti == time_d['info']['frames'][i]['events'][j]['killerId'] :
                    kill += 1
                if parti == time_d['info']['frames'][i]['events'][j]['victimId'] :
                    death += 1
            # WARD_PLACED
            if 'WARD_PLACED' in time_d['info']['frames'][i]['events'][j]['type'] :
                if parti == time_d['info']['frames'][i]['events'][j]['creatorId'] :
                    ward_placed += 1
            # SKILL_TREE
            if 'SKILL_LEVEL_UP' in time_d['info']['frames'][i]['events'][j]['type'] :
                if parti == time_d['info']['frames'][i]['events'][j]['participantId'] :
                    if skill_tree == '' :
                        skill_tree += str(time_d['info']['frames'][i]['events'][j]['skillSlot'])
                    else :
                        skill_tree += '|' + str(time_d['info']['frames'][i]['events'][j]['skillSlot'])
            # TOWER_KILL
            if 'BUILDING_KILL' in time_d['info']['frames'][i]['events'][j]['type'] :
                if time_d['info']['frames'][i]['events'][j]['killerId'] == parti :
                    tower_kill += 1
            # ITEM_ID
            if 'ITEM_PURCHASED' in time_d['info']['frames'][i]['events'][j]['type'] :
                if time_d['info']['frames'][i]['events'][j]['participantId'] == parti :
                    if item_id == '' :
                        item_id += str(time_d['info']['frames'][i]['events'][j]['itemId'])
                    else :
                        item_id += '|' +  str(time_d['info']['frames'][i]['events'][j]['itemId'])

        # TOTAL_GOLD
        total_gold = time_d['info']['frames'][i]['participantFrames'][str(parti)]['totalGold']

        # POSITION
        position = str(time_d['info']['frames'][i]['participantFrames'][str(parti)]['position']['x']) + '|'
        position += str(time_d['info']['frames'][i]['participantFrames'][str(parti)]['position']['y'])

        # GAME RESULT
        if mini_d['info']['participants'][parti-1]['win'] == True :
            win = 1
        else :
            win = 0

        # MATCH_ID
        match_id = time_data['metadata']['matchId']

        # MATCH_ID_MINI
        match_id_mini = mini_data['metadata']['matchId']


        if death != 0 :
            kda = round((kill + assist) / death, 2)
        else : kda = (kill + assist)

        lst_min_parti.append(parti)
        if match_id == match_id_mini :
            lst_min_parti.append(match_id)
        print(match_id)
        lst_min_parti.append(mini_t_time)
        lst_min_parti.append(mini_t_level)
        lst_min_parti.append(kda)
        lst_min_parti.append(skill_tree)
        lst_min_parti.append(tower_kill)
        lst_min_parti.append(ward_placed)
        lst_min_parti.append(total_gold)
        lst_min_parti.append(position)
        lst_min_parti.append(win)
        lst_min_parti.append(item_id)
        lst_min_parti.append(kill)
        lst_min_parti.append(assist)
        lst_min_parti.append(death)

        lst_parti.append(lst_min_parti)

    return lst_parti
```

```python
def final_time_line(mi_d, ti_d) :
    lst_ori = []
    for i in range(len(ti_d)) :
        t_d = ti_d[i]
        m_d = mi_d[i]

        for j in range(1, 11) :
            lst_ex = time_line_data(j, t_d, m_d)
            lst_ori.extend(lst_ex)

    timeline_columns = ['mini_t_participant', 'match_id', 'mini_t_time', 'mini_t_level', 'mini_t_kda', 'mini_t_skill_slot', 'mini_t_tower_kill', 'mini_t_ward', 'mini_t_total_gold', 'mini_position', 'minigame_result', 'item_id', 'kill', 'assist', 'death']

    df = pd.DataFrame(lst_ori, columns=timeline_columns)

    return df
```
